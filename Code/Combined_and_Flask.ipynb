```python
import pandas as pd
from sqlalchemy import create_engine
from sqlalchemy_utils import database_exists, create_database
```

```python
pd.set_option('display.max_rows',300)
```

# Population Data

```python
population_csv = "../Resources/Cleaned_CSV/Population_Data.csv"
population = pd.read_csv(population_csv)
population
```

# GDP Data

```python
gdp_csv = "../Resources/Cleaned_CSV/GDP_Data.csv"
gdp = pd.read_csv(gdp_csv)
gdp
```

# Total Emission Data

```python
total_emission_csv = "../Resources/Cleaned_CSV/Total_Emission_Data.csv"
total_emission = pd.read_csv(total_emission_csv)
total_emission
```

# Gas Emission Data

```python
gas_emission_csv = "../Resources/Cleaned_CSV/Gas_Emission_Data.csv"
gas_emission = pd.read_csv(gas_emission_csv)
gas_emission
```

# Liquid Emission Data

```python
liquid_emission_csv = "../Resources/Cleaned_CSV/Liquid_Emission_Data.csv"
liquid_emission = pd.read_csv(liquid_emission_csv)
liquid_emission
```

# Solid Emission Data

```python
solid_emission_csv = "../Resources/Cleaned_CSV/Solid_Emission_Data.csv"
solid_emission = pd.read_csv(solid_emission_csv)
solid_emission
```

# Combining Dataframes

```python
joined = pd.concat([population,gdp,total_emission,gas_emission,liquid_emission,solid_emission])
joined.head()
```

```python
joined = joined.fillna(value='N/A')
joined
```

```python
joined.to_csv("../Resources/Final_CSV/Data.csv", index=False)
```

# Database to SQL

```python
protocol = 'postgresql'
username = 'postgres'
password = 'postgres'
host = 'localhost'
port = 5432
database_name = 'World_Data'
rds_connection_string = f'{protocol}://{username}:{password}@{host}:{port}/{database_name}'
engine = create_engine(rds_connection_string)
if not database_exists(engine.url):
    create_database(engine.url)
```

```python
engine.table_names()
```

```python
joined.to_sql(name='World_Data', con=engine, if_exists='append', index=False)
```
