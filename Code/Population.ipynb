# Country Codes

```python
#import countries code csv
code_csv = "../Resources/CountryCodes.csv"
code = pd.read_csv(code_csv)
```

```python
code.columns
```

```python
code = code[['CLDR display name', 'ISO3166-1-Alpha-2', 'ISO3166-1-Alpha-3']]
code
```

# Match Country Codes

```python
country_three = country.merge(code, how='left', left_on='Country Code', right_on='ISO3166-1-Alpha-2', sort=True)
```

```python
country_three = country_three.drop(['CLDR display name','Country Code','ISO3166-1-Alpha-2'],axis=1)
```

```python
country_three = country_three.rename(columns={'ISO3166-1-Alpha-3':'Country Code'})
country_three
```

```python
# joined = pd.concat([population,country],join='inner',ignore_index=True)
# joined = country.join(population, how='left', on='Country')

joined = country_three.merge(population, how='left', left_on='Country Code', right_on='Country Code', sort=True)
joined
```

```python
joined = joined.dropna(how='any')
joined
```

```python
for x in range(1960,2021):
    joined = joined.rename(columns={f'{x} [YR{x}]':f'{x}'})
joined
```

```python
joined = joined.drop(['1960', '1961', '1962',
       '1963', '1964', '1965', '1966', '1967', '1968', '1969', '1970', '1971',
       '1972', '1973', '1974', '1975', '1976', '1977', '1978', '1979', '1980',
       '1981', '1982', '1983', '1984', '1985', '1986', '1987', '1988', '1989'],axis=1)
joined
```
