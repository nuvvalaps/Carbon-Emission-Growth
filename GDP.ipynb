# GDP Data

```python
#import population csv
gdp_csv = "../Resources/GDP_Data.csv"
gdp = pd.read_csv(gdp_csv)
```

```python
gdp = gdp.dropna(how='any')
gdp = gdp.drop(['Series Name','Series Code','2021 [YR2021]','2020 [YR2020]','Country Name'],axis=1)
```

# Country Data

```python
#import countries csv
country_csv = "../Resources/LatLong.csv"
country = pd.read_csv(country_csv)
```

```python
country = country.dropna(how='any')
```

```python
country.dtypes
```

```python
country.head()
```

# Country Codes

```python
#import countries code csv
code_csv = "../Resources/CountryCodes.csv"
code = pd.read_csv(code_csv)
```

```python
code.columns
```

```python
code = code[['CLDR display name', 'ISO3166-1-Alpha-2', 'ISO3166-1-Alpha-3']]
```

# Match Country Codes

```python
country_three = country.merge(code, how='left', left_on='Country Code', right_on='ISO3166-1-Alpha-2', sort=True)
```

```python
country_three = country_three.drop(['CLDR display name','Country Code'],axis=1)
```

```python
country_three = country_three.rename(columns={'ISO3166-1-Alpha-3':'Country Code'})
country_three.head()
```

```python
joined = country_three.merge(gdp, how='left', left_on='Country Code', right_on='Country Code', sort=True)
```

```python
joined = joined.dropna(how='any')
```

```python
for x in range(1960,2020):
    joined = joined.rename(columns={f'{x} [YR{x}]':f'{x}'})
```

```python
# joined.columns
```

```python
joined = joined.drop(['1960', '1961', '1962',
       '1963', '1964', '1965', '1966', '1967', '1968', '1969', '1970', '1971',
       '1972', '1973', '1974', '1975', '1976', '1977', '1978', '1979', '1980',
       '1981', '1982', '1983', '1984', '1985', '1986', '1987', '1988', '1989'],axis=1)
```

```python
# joined = joined.drop(['Lat','Long'],axis=1)
```

```python
joined = joined.replace('..','0')
joined
```

```python
# joined['Series Name'] = 'GDP'
joined.head()
```

```python
joined = joined.drop(['Country Code'],axis=1)
```

```python
joined = joined.rename(columns={'ISO3166-1-Alpha-2':'Country Code'})
joined.head()
```
